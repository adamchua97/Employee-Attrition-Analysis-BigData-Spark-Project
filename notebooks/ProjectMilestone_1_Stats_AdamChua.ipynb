# Histograms

### Create separate df for numerical and categorical variables

```python
numerical_cols = df.select_dtypes(include=['number']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
```

```python
numerical_cols
```

```python
categorical_cols
```

```python
# separate dataframes

numerical_df = df[numerical_cols]
categorical_df = df[categorical_cols]
```

```python
numerical_df.head()
```

```python
categorical_df.head()
```

## Numerical Variables

```python
numerical_df.info()
```

```python
# Create a figure and a grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plot histograms on each subplot
sns.histplot(numerical_df, x='Age', ax=axes[0, 0])
axes[0, 0].set_title('Age')

sns.histplot(numerical_df, x='Job_Level', ax=axes[0, 1])
axes[0, 1].set_title('Job Level')

sns.histplot(numerical_df, x='Monthly_Income', ax=axes[1, 0])
axes[1, 0].set_title('Monthly Income')

sns.histplot(numerical_df, x='Hourly_Rate', ax=axes[1, 1])
axes[1, 1].set_title('Hourly Rate')

# Adjust the layout
plt.tight_layout()
plt.show()
```

## Categorical Variables

```python
categorical_df.info()
```

```python
# Categorical variable counts

for col in categorical_cols:
    print(categorical_df[col].value_counts(), "\n")
```

```python
# Create a figure and a grid of subplots
fig, axes = plt.subplots(3, 2, figsize=(12, 14))

# Plot histograms on each subplot
sns.barplot(data=categorical_df['Gender'].value_counts(), ax=axes[0, 0], width=0.5)
axes[0, 0].set_title('Gender Frequency Bar Chart')

sns.barplot(data=categorical_df['Marital_Status'].value_counts(), ax=axes[0, 1], width=0.5)
axes[0, 1].set_title('Marital Status Frequency Bar Chart')

sns.barplot(data=categorical_df['Department'].value_counts(), color='orange', ax=axes[1, 0], width=0.5)
axes[1, 0].set_title('Department Frequency Bar Chart')

sns.barplot(data=categorical_df['Job_Role'].value_counts(), color='orange', ax=axes[1, 1], width=0.5)
axes[1, 1].set_title('Job Role Frequency Bar Chart')

sns.barplot(data=categorical_df['Overtime'].value_counts(), color='green', ax=axes[2, 0], width=0.5)
axes[2, 0].set_title('Overtime Frequency Bar Chart')

sns.barplot(data=categorical_df['Attrition'].value_counts(), color='green', ax=axes[2, 1], width=0.5)
axes[2, 1].set_title('Attrition Frequency Bar Chart')

# Adjust the layout
plt.tight_layout()
plt.show()
```
